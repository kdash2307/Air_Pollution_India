```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#importing dataset
dataset=pd.read_csv("G:\Extras\Computer\Internship\data.csv",encoding='ISO-8859-1',index_col='state')
```

```python
#Get basic info about the data type and no. of null values and all
dataset.info()
```

```python
#categorise common values like state and location to reduce memory usage
dataset.index=dataset.index.astype('category')
```

```python
dataset['location']=dataset['location'].astype('category')
```

```python
dataset['type']=dataset['type'].astype('category')
```

```python
#dropping unneccesary columns like agency,sampling date and stn_code
dataset.drop(columns=['stn_code','agency','sampling_date','location_monitoring_station','date'],inplace=True)
dataset
```

```python
#this helps us to know number of nan values in every column
len(dataset) - dataset.count()

```

```python
#we observe a large number of nan values in psm2_5 and spm so we are dropping
dataset.drop(columns=['pm2_5','spm'],inplace=True)
```

```python
dataset[dataset['location'].isna()]
```

```python
#we drop Andaman,Lakshwadeep and Tripura too as they dont have any required values
dataset.dropna(how='all',inplace=True)
```

```python
#we group by the location and remove the missing values
```

```python
grouped=dataset.groupby(['state','location','type']).first()
```
